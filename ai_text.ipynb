```python
from transformers import pipeline
def ai_text_generator(prompt):

    generator = pipeline("text-generation", model="gpt2")

    outputs = generator(prompt, max_length=150, num_return_sequences=1)

    return outputs[0]['generated_text']

if __name__ == "__main__":
    print("=== AI Text Generator ===")
    while True:
        user_prompt = input("\nEnter your prompt (or type 'exit' to quit): ")
        if user_prompt.lower() == "exit":
            print("Goodbye!")
            break
        result = ai_text_generator(user_prompt)
        print("\n--- Generated Text ---\n")
        print(result)
```
